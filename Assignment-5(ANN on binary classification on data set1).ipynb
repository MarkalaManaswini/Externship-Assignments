```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("diabetes.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().any()
```

```python
x=dataset.iloc[:,0:8].values
y=dataset.iloc[:,8:9].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1000)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
cat=Sequential()
```

```python
cat.add(Dense(8,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(16,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(1,kernel_initializer="random_uniform",activation="sigmoid"))
```

```python
cat.compile("adam","binary_crossentropy",metrics=["accuracy"])
```

```python
cat.fit(x_train,y_train,batch_size=35,epochs=50)
```

```python
ypred=cat.predict(x_test)
```
